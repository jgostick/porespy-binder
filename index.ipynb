# Introduction to PoreSpy

PoreSpy is a collection of functions that operating on volumetric images represented as Numpy arrays.

There are several powerful image analysis packages in the Scientific Python ecosystem, including ``scipy.ndimage`` and ``scikit-image``.  These offer general purpose functions such as binary morphology and distance transforms.  

PoreSpy uses the building blocks provided by these libraries to create functions which are specifically for analyzing images of porous materials.  

PoreSpy is organized into the following modules:

---
- [generators](https://porespy.org/modules/generated/porespy.generators.html#module-porespy.generators): Functions for generating artificial images useful for testing and validation
---
- [filters](https://porespy.org/modules/generated/porespy.filters.html): Functions for manipulating images to replace voxel values with some computed values
---
- [metrics](https://porespy.org/modules/generated/porespy.metrics.html): Functions for analyzing images to extract quantitative information
---
- [simulations](https://porespy.org/modules/generated/porespy.simulations.html): Functions for performing physical simulations directly on images
---
- [networks](https://porespy.org/modules/generated/porespy.networks.html): Functions for extracting pore networks from images for use in OpenPNM
---
- [visualization](https://porespy.org/modules/generated/porespy.visualization.html): Functions for performing basic visualizations of 3D images
---
- [io](https://porespy.org/modules/generated/porespy.io.html): A small set of functions for converting between various formats
---
- [tools](https://porespy.org/modules/generated/porespy.tools.html): Functions mostly used within PoreSpy itself
---
